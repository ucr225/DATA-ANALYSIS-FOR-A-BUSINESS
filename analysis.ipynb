```python
# Importing Libraries
import pandas as pd
#from datasets import load_dataset
import matplotlib.pyplot as plt

try:
    import pandas as pd
    print("Pandas está instalado")
except ImportError:
    print("Pandas no está instalado")
```

```python
df = pd.read_csv('sales.csv')
df.head()
#validando que este bien cargado el archivo
```

```python
df.info()
#getting info from the dataset
```

```python
total_sales = df['Cantidad de Pares'].sum() 
print(total_sales) 
print('Cantidad de Pares vendidos: ', total_sales)
print(df['Cantidad de Pares'].mean())
print('Promedio de pares vendidos: ', df['Cantidad de Pares'].mean())
```

```python
# Filtrar registros que tienen género 'F'
df_female = df.loc[df['Genero'] == 'F']

print(df_female)
```

```python
# Filtrar registros que tienen género 'F'
df_male = df.loc[df['Genero'] == 'M']

print(df_male)
```

```python
# Obtener el número de filas del DataFrame
rows_male = df_male.shape[0]
print(rows_male)
```

```python
rows_female = df_female.shape[0]
print(rows_female)
```

```python
# Crear un gráfico circular (pastel) de la columna 'genero'
table = df['Genero'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Porcentaje de clientes en función de su género')
plt.ylabel('')
# Mostrar la gráfica
plt.show()
```

```python
total_women=df_female['Cantidad de Pares'].sum() 
print(total_women)
```

```python
total_men=df_male['Cantidad de Pares'].sum() 
print(total_men)

```

```python
sales_comparing  = [total_men, total_women]
labels = ['Men', 'Women']
# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sales_comparing, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales')
plt.axis('equal')  # Para asegurar que el gráfico de pastel sea un círculo
plt.show()
```

```python
edad_male =df_male['Edad'].mean()
print(edad_male, 'edad promedio de hombres')
# Crear un gráfico de barras de la columna 'Edad'
table = df_male['Edad'].value_counts().plot(kind='bar')
plt.title('Distribución de la edad de los clientes masculinos')
plt.xlabel('Edad')
plt.ylabel('Número de clientes')
# Mostrar la gráfica
plt.show()
```

```python
edad_female =df_female['Edad'].mean()
print(edad_female, 'edad promedio de mujeres')

# Crear un gráfico de barras de la columna 'Edad'
table = df_female['Edad'].value_counts().plot(kind='bar')
plt.title('Distribución de la edad de los clientes femeninos')
plt.xlabel('Edad')
plt.ylabel('Número de clientes')
# Mostrar la gráfica
plt.show()
```

```python

tipo_de_zapatos = df_male['Tipo de Zapatos'].value_counts()

total = tipo_de_zapatos.sum()

plt.figure(figsize=(10, 6))
bars = plt.bar(tipo_de_zapatos.index, tipo_de_zapatos.values, color='#C1121F')
plt.title('Distribución de los Tipos de Zapatos')
plt.xlabel('Tipo de Zapatos')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)  # Para rotar las etiquetas del eje x si son largas

# Añadir porcentajes sobre cada barra
for bar in bars:
    height = bar.get_height()
    percentage = f'{height / total * 100:.1f}%'
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, percentage, ha='center', va='bottom')

plt.show()
```

```python
tipo_de_zapatos = df_female['Tipo de Zapatos'].value_counts()

total = tipo_de_zapatos.sum()

plt.figure(figsize=(10, 6))
bars = plt.bar(tipo_de_zapatos.index, tipo_de_zapatos.values, color='purple')
plt.title('Distribución de los Tipos de Zapatos')
plt.xlabel('Tipo de Zapatos')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)  # Para rotar las etiquetas del eje x si son largas

# Añadir porcentajes sobre cada barra
for bar in bars:
    height = bar.get_height()
    percentage = f'{height / total * 100:.1f}%'
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, percentage, ha='center', va='bottom')

plt.show()
```
